üîπ 4. **Feature Engineering**

üéØ Objective:
Prepare your dataset so that machine learning models can understand and learn from it effectively.

‚úÖ Step 1: Define Features and Target

```python
features = [
    'Use of insecticide-treated bed nets (% of under-5 population)',
    'Children with fever receiving antimalarial drugs (% of children under age 5 with fever)',
    'Intermittent preventive treatment (IPT) of malaria in pregnancy (% of pregnant women)',
    'People using safely managed drinking water services (% of population)',
    'People using safely managed sanitation services (% of population)',
    'Malaria cases reported',
    'Year'
]

target = 'Incidence of malaria (per 1,000 population at risk)'
```

‚úÖ Step 2: Encode Country Name 

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
east_africa_df['Country_Code'] = encoder.fit_transform(east_africa_df['Country Name'])

# Add Country_Code to features
features.append('Country_Code')
```

‚úÖ Step 3: Scale Features

```python
from sklearn.preprocessing import StandardScaler

X = df[features]
y = df[target]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

üîπ 5. **TRAIN/TEST SPLIT**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=100
)
```

```python
print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

üîπ 6. **MODEL BUILDING**

START WITH SIMPLE MODELS, THEN TUNE AND COMPARE

üìò (i) **LINEAR REGRESSION:**

‚úÖ Step 1:Train the Model (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression ()
lr.fit(X_train,y_train)
```

‚úÖ Step 2:Applying the model to make Prediction

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
print(y_lr_train_pred)
print(y_lr_test_pred)
```

‚úÖ Step 3: Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2  = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
```

```python
import pandas as pd

# Create the result DataFrame correctly
lr_results = pd.DataFrame([['LinearRegression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]],
                          columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])

# Display the result
print(lr_results)
 
```

üå≥ (ii) **RANDOM FOREST:**

‚úÖ Step 1:Train a Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the model
rf = RandomForestRegressor(max_depth=2, random_state =100)

# Train the model
rf.fit(X_train, y_train)
```

‚úÖ Step 2:Applying the model to make Prediction

```python
# Predictions using Random Forest
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)

# View predictions
print("Random Forest - Training Predictions:")
print(y_rf_train_pred)

print("\nRandom Forest - Test Predictions:")
print(y_rf_test_pred)
```

‚úÖ Step 3: Evaluate Model Performance

üìò Random Forest Performance Evaluation

```python
# Evaluate Random Forest performance
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
import pandas as pd

# Create the result DataFrame for Random Forest
rf_results = pd.DataFrame([['RandomForest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]],
                          columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])

# Display the result
print(rf_results)
```

(iii) **Decision Tree for Malaria Incidence Prediction**


‚úÖ Step 1:Train a  Decision Tree Model

```python
from sklearn.tree import DecisionTreeRegressor
# Initialize the model
dt = DecisionTreeRegressor(random_state=42)
# Train the model
dt.fit(X_train, y_train)
```

‚úÖ Step 2:Applying the model to make Prediction

```python
y_dt_train_pred = dt.predict(X_train)
y_dt_test_pred = dt.predict(X_test)
```

‚úÖ Step 3: Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

dt_train_mse = mean_squared_error(y_train, y_dt_train_pred)
dt_train_r2 = r2_score(y_train, y_dt_train_pred)

dt_test_mse = mean_squared_error(y_test, y_dt_test_pred)
dt_test_r2 = r2_score(y_test, y_dt_test_pred)
```

```python
import pandas as pd

dt_results = pd.DataFrame([['DecisionTree', dt_train_mse, dt_train_r2, dt_test_mse, dt_test_r2]],
                          columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])

print(dt_results)
```

**MODEL COMPARISON**

Let's summarize and compare the three models ‚Äî Linear Regression, Random Forest, and Decision Tree ‚Äî in a clean table based on the evaluation metrics:

```python
import pandas as pd

# Combine results
comparison_df = pd.DataFrame([
    ['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2],
    ['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2],
    ['Decision Tree', dt_train_mse, dt_train_r2, dt_test_mse, dt_test_r2]
], columns=['Model', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2'])

# Display
print(comparison_df)
```

üìä **What to Look for in Model Evaluation**

Lower MSE

‚úÖ Mean Squared Error (MSE) measures how close the predicted values are to the actual values.

üîΩ Lower values indicate better model accuracy.

Higher R¬≤

‚úÖ R-squared (R¬≤) shows how much of the variation in the target variable is explained by the model.

üîº Higher values (closer to 1) indicate the model explains more variance.

Train vs Test Scores

‚ö†Ô∏è If the test scores are much worse than the train scores, it may indicate overfitting ‚Äî the model learned the training data too well and performs poorly on new data.

Look for a balance between training and test performance.



üìä **Model Comparison Bar Chart (MSE and R¬≤)**

```python
plt.figure(figsize=(14, 6))

# Test MSE Comparison
plt.subplot(1, 2, 1)
sns.barplot(data=comparison_df, x='Model', y='Test MSE', palette='Blues_d')
plt.title('Model Comparison: Test MSE')
plt.ylabel('Mean Squared Error')

# Test R2 Comparison (updated column name)
plt.subplot(1, 2, 2)
sns.barplot(data=comparison_df, x='Model', y='Test R2', palette='Greens_d')  # e.g., if column is 'R2'
plt.title('Model Comparison: Test R¬≤')
plt.ylabel('R-squared Score')

plt.tight_layout()
plt.show()
```

## üß† How the Line Plot Achieves Insightful Evaluation

### 1. Tracking Closest to Actual (Model Accuracy)
**What to look for:**

The model line (e.g., ‚ÄúRandom Forest‚Äù) that stays **closest** to the ‚ÄúActual‚Äù line (ground truth) across all years.

**What it means:**

- If a model‚Äôs prediction line nearly overlaps the actual values, it‚Äôs **very accurate**.
- **Example:** If Random Forest‚Äôs line hugs the actual line across the timeline, it's likely the **best model**.

---

### 2. Spotting Over- or Under-Prediction
**What to look for:**

- **Over-prediction** ‚Üí model's line is consistently **above** the actual line.
- **Under-prediction** ‚Üí model's line is consistently **below** the actual line.

**What it means:**

- A pattern of over- or under-predicting reveals **bias** in the model.
- **Example:** If the Linear Regression model keeps **overestimating** in early years, it might not be capturing some **temporal patterns**.

---

### 3. Evaluating Trend Matching (Pattern Fit)
**What to look for:**

Does the shape of the model‚Äôs line follow the same **ups and downs** as the actual line?

**What it means:**

- A model that correctly predicts **trend directions** (e.g., increasing/decreasing years) shows it understands the **temporal structure**.
- Even if exact numbers are a bit off, **capturing trends** is important in forecasting.
